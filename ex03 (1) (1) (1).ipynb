```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1. Read the Image (Color)
# -----------------------------
img = cv2.imread('input0.jpg')

if img is None:
    print("Image not found. Keep input4.png in same folder.")
    exit()

# Convert BGR to RGB for matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rows, cols, ch = img_rgb.shape

# -----------------------------
# 2. Translation
# -----------------------------
tx, ty = 100, 50
translation_matrix = np.float32([[1, 0, tx],
                                 [0, 1, ty]])

translated = cv2.warpAffine(img_rgb, translation_matrix, (cols, rows))

# -----------------------------
# 3. Scaling
# -----------------------------
scaled = cv2.resize(img_rgb, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)

# -----------------------------
# 4. Rotation
# -----------------------------
angle = 45
rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
rotated = cv2.warpAffine(img_rgb, rotation_matrix, (cols, rows))

# -----------------------------
# 5. Reflection
# -----------------------------
# 1 = Horizontal flip
reflected = cv2.flip(img_rgb, 1)

# -----------------------------
# 6. Display Results
# -----------------------------
plt.figure(figsize=(12,10))

plt.subplot(3,2,1)
plt.imshow(img_rgb)
plt.title("Original Color Image")
plt.axis("on")

plt.subplot(3,2,2)
plt.imshow(translated)
plt.title("Translated Image")
plt.axis("off")

plt.subplot(3,2,3)
plt.imshow(scaled)
plt.title("Scaled Image")
plt.axis("on")

plt.subplot(3,2,4)
plt.imshow(rotated)
plt.title("Rotated Image (45)")
plt.axis("off")

plt.subplot(3,2,5)
plt.imshow(reflected)
plt.title("Reflected Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
